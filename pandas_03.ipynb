```python
import pandas as pd

student = {'Name':['Atul','Dhanush','Kraim','Ajay'],
           'Age': [29,30,28,25],
           'Status': ['Student','JOB','Student','Umemployed']}

# Lets first create 3 more columns so that we can delete them
df = pd.DataFrame(student)
df
```

```python
# We will add a row

anotherStudent = ['Raj', 28, 'Student']

df.loc[len(df.index)] = anotherStudent
df
```

```python
# Lets first create 3 more columns so that we can delete them

busy = ['Yes', 'No','Yes','No','Yes']

birth_years = ['1995', '1994', '1993', '1992', '1991']

born_cities= ['Dhaka','Pakistan','Chittagong','Dhaka','Delhi']

df['Busy'] = busy
df['Birth Year'] = birth_years
df['Born City'] = born_cities

df
```

```python
# Let's delete a row

df.drop(len(df)-1, axis=0,inplace=True)
df
```

```python
# Lets delete column

df.drop(['Birth Year','Born City'], axis=1, inplace=True)
df
```

```python
# Lets rename columns

df.rename(columns={'Name':'First Name','Age':'Age in Years'}, inplace=True)
df
```

```python
# Let'S rename the index

df.rename(index={0:'A',1:'B',2:'C',3:'D'}, inplace=True)
df
```
