```python
import pandas as pd
```

```python
student = {'Name':['Atul','Dhanush','Kraim','Ajay'],
           'Age': [29,30,28,25],
           'Status': ['Student','JOB','Student','Umemployed']}

df = pd.DataFrame(student)
print(df)
```

### Now Lets use set_index() method to use our colums as index instead of 0-3

```python
df.set_index('Name',inplace=True)
print(df)
```

```python
#Lets access using Name colums

print(df.loc['Atul'],"\n")
print(df.loc['Dhanush'],"\n")
print(df.loc['Kraim'],"\n")
print(df.loc['Ajay'])
```

```python
# we can put specific index using range index
student = {'Name':['Atul','Dhanush','Kraim','Ajay'],
           'Age': [29,30,28,25],
           'Status': ['Student','JOB','Student','Umemployed']}

df = pd.DataFrame(student)
```

```python
df
```

```python
df = pd.DataFrame(student,index=pd.RangeIndex(100,104))
df
```

## Array

```python
# lets create a python list

nums = [x for x in range(2,9,2)]

# the pandas array from this python list

nums_arr = pd.array(nums)
print(nums_arr)
```

```python
# lets do this by sending python list as
# an argument in the pandas array

pandas_arr = pd.array([x for x in range(2,9,2)])
print(pandas_arr)
```

```python
# We can specify data types when creating array

pandas_arr = pd.array([x for x in range(2,9,2)],dtype=int)
print(pandas_arr)

float_arr = pd.array([1.1,1.2,1.5,1.9],dtype='float')
print(float_arr)
```

##DataFrame Manipulation of Rows/ Columns



1.   Add
2.   remove
3.   Rename



```python
# Lets again create our student dataset

student = {'Name':['Atul','Dhanush','Kraim','Ajay'],
           'Age': [29,30,28,25],
           'Status': ['Student','JOB','Student','Umemployed']}

df = pd.DataFrame(student)

df
```

```python
# We will add another column as city

city = ['Chemnitz','Leipzig', 'Berlin','Hamburg']

df['City'] = city

df
```

```python
# We will add a row

anotherStudent = ['Raj', 28, 'Student', 'Chemnitz']

df.loc[len(df.index)] = anotherStudent

df
```

```python
# Lets first create 3 more columns so that we can delete them

busy = ['Yes', 'No','Yes','No','Yes']

birth_years = ['1995', '1994', '1993', '1992', '1991']

born_cities= ['Dhaka','Pakistan','Chittagong','Dhaka','Delhi']

df['Busy'] = busy
df['Birth Year'] = birth_years
df['Born City'] = born_cities

df
```
