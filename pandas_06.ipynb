```python
import pandas as pd

# Creating a sample student dataset
data = {
    'Name': ['Atul', 'Ajay', 'Arafat', 'Dhanush', 'Karim'],
    'Age': [18, 19, 17, 18, 20],
    'Grade': [85, 92, 78, 95, 88],
    'Subject': ['Math', 'Science', 'English', 'History', 'Math']
}

# Creating the DataFrame
student = pd.DataFrame(data)
```

```python
student[student['Age'] > 18]
```

```python
# Multiple Conditions

student[(student['Age'] > 18) & (student['Grade'] > 90)]
```

```python
# isin used used to check for specific value
student[student['Name'].isin(['Atul'])]
```

```python
student[student['Age'].notnull()] ## Check For Non-Null Value
```

```python
student[student['Age'].isnull() == False] ## Check For Non-Null Value
```
