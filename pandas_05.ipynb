## Series

```python
import pandas as pd

# Lets create a series

series_1 = pd.Series([10,20,70])

print(series_1)
```

```python
# Series with custom index

series_2 = pd.Series([10,20,70], index=['a','b','c'])

print(series_2)
```

```python
items = {'Dog':'Happy','Cat': 'SAD','COW':'FUNNY'}

# Python Dictionary to Series
series_3 = pd.Series(items)

print(series_3)
```

```python
# Empty Seies

empty_series = pd.Series(dtype=float)

print(empty_series)
```

## DataFrames

```python
students = {'Name':['Sakib','Atul','Rahman'], 'Age':[21,23,21]}

df = pd.DataFrame(students)

print(df)
```

```python
# Custom index

df = pd.DataFrame(students, index=['s1','s2','s3'])

print(df)
```
