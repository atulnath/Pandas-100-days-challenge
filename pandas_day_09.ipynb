```python
import pandas as pd
import numpy as np
```

```python

data = {
    'name':['Atul', 'Rahim', 'Ali', 'Dahnush', 'Rahman'],
    'age': [25,29,27,31,20],
    'city': ['Dhaka', 'Berlin', 'Chiitagong', 'Munich','Dresden']
}

df = pd.DataFrame(data)
df
```

## Modifications

```python
salary =[2000,1800,2200,1100,2000]
df['salary'] = salary
df
```

```python
df['salary'] = df['salary'] * 2.5 # 25% increase

df
```

```python
df['salary'] = df['salary'] / 2.5 # 25% increase

df
```

```python
df['salary'] =df['salary'] + df['salary'] * 0.25 # 25% increase

df
```

```python
# Let add bonus based on salary 5%

df['bonus'] = df['salary'] * 0.05
df
```

```python
df['total_salary'] = df['salary'] + df['bonus']
df
```
