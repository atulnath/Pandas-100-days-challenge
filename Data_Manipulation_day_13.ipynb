```python
import pandas as pd

data = {
    'Name': ['Atul', 'Dhanush', 'Karim', 'Hassan', 'Robi'],
    'Age': [25, 30, 22, 28, 35],
    'Salary': [30000, 45000, 25000, 40000, 50000]
}

df = pd.DataFrame(data)

print(df)
```

```python
## Filtering
df[df['Age'] > 20]
```

```python
df[df['Age'] > 30]
```

```python
sorted_with_salary = df.sort_values(by='Salary', ascending=False)
print(sorted_with_salary)
```

```python
df[ ['Age']]
```

```python
df['Annual Salary'] = df['Salary'] * 12
print(df)
```

```python
df['Department'] = ['IT', 'HR', 'IT', 'HR', 'IT']
print(df)
```

```python
# Filtering multiple conditions

df[(df['Age'] > 25) & (df['Salary'] > 35000)]
```

```python
sorted_age_salary = df.sort_values(by=['Age', 'Salary'], ascending=[False, True])
print(sorted_age_salary)
```

```python
df[['Age','Name','Salary']]
```

```python
df['Bonus'] = df['Salary'] * 0.10
```

```python
df
```

```python
df[df['Department'] == 'IT']
```

```python
def Age_Class(age):
  if age < 25:
    return 'Young'
  elif age < 35:
    return 'Middle Age'
  else:
    return 'Senior'
df['Age Class'] = df['Age'].apply(Age_Class)
```

```python
df
```

```python
df.groupby('Department').agg({'Salary': 'mean', 'Age': 'max'})
```

```python
df.drop(columns=['Annual Salary'])
```

```python
df['Salary Class'] = df['Salary'].apply(lambda x: 'High' if x > 40000 else 'Low')
df
```
