```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['Rahim', 'Karim', 'Rahim', 'Sadia', None, 'karim', 'Ayesha', 'ayesha', np.nan, 'Rahim'],
    'Age': [25, np.nan, 25, 30, 28, np.nan, 22, 22, 35, 25],
    'Salary': [50000, 60000, 50000, np.nan, 45000, 60000, 40000, 40000, np.nan, 50000],
    'Join_Date': ['2021-01-15', '2020-05-10', '2021-01-15', '2022-03-22', '2021-12-01',
                  '2020-05-10', '2023-01-10', '2023/01/10', '2022-06-15', '2021.01.15'],
    'Department': ['HR', 'IT', 'HR', 'Finance', 'finance', 'IT', 'Marketing', 'marketing', 'HR', 'hr']
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Information about data

df.info()
```

### From the df.info() we can read

1.  Name, Age, Salary column has 8 non-null entry
2. total 2 float64 and 3 object, could be date and string

```python
# Lets Count the non-null values
df.isnull().sum()
```

```python
# Im the name column we can see there is 2 missing value
# Lets fill it with Unknown

df['Name'] = df['Name'].fillna('Unknown')
```

```python
df
```

```python
# Now we can see the missing value in both Age and Salary

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

```python
df
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
```

```python
df
```

```python
# Lets check we have duplicate row or not

df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df = df.drop_duplicates()
```
