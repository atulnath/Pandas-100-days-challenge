```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'name': ['Rahim', 'Karim', 'Sonia', 'Mina', 'Rashed', 'Ayesha', 'Fahim', 'Nadia'],
    'age': [25, 30, 22, 35, 28, 27, 32, 29],
    'city': ['Dhaka', 'Chittagong', 'Khulna', 'Dhaka', 'Rajshahi', 'Sylhet', 'Dhaka', 'Barisal'],
    'salary': [30000, 45000, 28000, 50000, 32000, 40000, 47000, 35000],
    'department': ['HR', 'IT', 'Marketing', 'Finance', 'IT', 'HR', 'Finance', 'Marketing']
})
```

```python
df.head()
```

```python
# Basic Filtering: Age > 30
df[df['age'] > 30]
```

```python
df[df['age'] > 25]
```

```python
df[df['age'] > 27]
```

```python
# Single column filtering: city = Chittagong
df[df['city'] == 'Chittagong']
```

```python
# Single column filtering: city = Chittagong
df[df['city'] == 'Dhaka']
```

```python
df[df['department'] == 'Finance']
```

```python
df[df['salary'] > 30000]
```

```python
# multicolumn filtering: age > 25 and salary > 40000

df[(df['age'] > 25) & (df['salary'] > 40000)]
```

```python
# Filter multiple values in single column: city: dhaka, chiitagong

df[df['city'].isin(['Dhaka', 'Chittagong'])]
```

```python
# Filter specific substring

df[df['name'].str.contains('im')]
```

```python
# Range filtering

df[df['salary'].between(30000, 45000)]
```

```python
# custom filter function

def custom_filter(row):
    return row['age'] > 25 and row['city'] == 'Dhaka'

df[df.apply(custom_filter, axis=1)]
```

```python
two_columns = df[['name', 'age']]
two_columns
```

```python
# nlargest, nsamlleset

large = df.nlargest(3, 'salary')
small= df.nsmallest(3, 'salary')

large, small
```
